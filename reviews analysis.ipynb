# Scraping

Определим функцию для скрэпинга отзывов с сайта turistplot.com

```python
def clean_string(column):
    return column.apply(lambda x: x.replace("\n",'',2)).apply(lambda x: x.replace('  ',''))

def scrape_reviews(PATH, n_pages, sleep_time = 0.3):
    """
    Scrape reviews from Turistpilot and save it in pd.DataFrame
    :param PATH: path to pages
    :n_pages: number of pages to scrap
    :returns: pd.DataFrame with header, review, rating, name, location, date
    """
    names = []
    ratings = []
    headers = []
    reviews = []
    dates = []
    locations = []
    
    for p in tqdm(range(1, n_pages)):
    
        sleep(sleep_time)
    
        http = requests.get(f'{PATH}{p}')
        bsoup = BeautifulSoup(http.text, 'html.parser')
        data_containers = bsoup.find_all('div', class_ = "styles_cardWrapper__LcCPA styles_show__HUXRb styles_reviewCard__9HxJJ")
    
        for container in data_containers:
            name = container.find(class_='typography_heading-xxs__QKBS8 typography_appearance-default__AAY17').text
            names.append(name)
    
            rating = int(container.find(class_='styles_reviewHeader__iU9Px')['data-service-review-rating'])
            ratings.append(rating)
            
            header = container.find(class_='typography_heading-s__f7029 typography_appearance-default__AAY17').text
            headers.append(header)
    
            try:
                review = container.find(class_='typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn').text
            except:
                review = None
            reviews.append(review)
    
            profile_ref = container.find(class_='link_internal__7XN06 link_wrapper__5ZJEx styles_consumerDetails__ZFieb')['href']
            profile_link = 'https://www.trustpilot.com'+ profile_ref
            user_profile = requests.get(f'{profile_link}')
            user_soup = BeautifulSoup(user_profile.text, 'html.parser')
            try:
                user_location = user_soup.find(class_ = 'typography_body-l__KUYFJ typography_appearance-subtle__8_H2l').text
            except:
                user_location = None
            locations.append(user_location)
    
            date = container.find(class_='typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_datesWrapper__RCEKH').time['datetime'][:10]
            dates.append(datetime.datetime.strptime(date, '%Y-%m-%d').date())
    
    rev_df = pd.DataFrame(list(zip(headers, reviews, ratings, names, locations, dates)),
                  columns = ['header','review','rating', 'name', 'location', 'date'])
    
    rev_df.review = clean_string(rev_df.review)
    rev_df.name = clean_string(rev_df.name)
    rev_df.location = clean_string(rev_df.location)
    rev_df.location = rev_df.location.apply(lambda x: x.split(',',1)[-1])
    rev_df.rating = rev_df.rating.astype('int')
    rev_df.date = pd.to_datetime(rev_df.date)
    
    return rev_df
```

```python
mm_reviews_count.describe()
```

Таким образом, мы имеем в среднем 59 отзывов каждый месяц, а среднеквадратичное отклонение 36. Это и подтверждает тот факт, что количество отзывов сильно колеблется. Заметим, что минимальное число отзывов равно 4. И заранее можем предположить, что это могло быть имеено в тот месяц, когда по рейтингу была достигнута оценка 2.0. Чтобы убедится в этом, сравним оба этих графика:

```python
monthly_data = pd.concat([
    (np.log(mm_rating)-np.log(mm_rating).mean())/np.log(mm_rating).std(), 
    (np.log(mm_reviews_count)-np.log(mm_reviews_count).mean())/np.log(mm_reviews_count).std()], axis=1)
monthly_data.head()
```

(для того, чтобы можно было наложить один график на другой, значения рейтинга и количества отзывов были прологарифмированы, а затем подвергнуты преобразованию стандартизации)

```python
monthly_data.plot(figsize=(12,5))
plt.title('comparing number of reviews with average rating by month')
plt.show()
```

Как можно видеть из полученного графика, наше предположение о совпадении упадка по количеству отзывов и средней оценке подтвердилось. Действительно, минимальные значения обеих величин достигаются в одном месяце. Таким образом, можно сделать вывод, что низкий рейтинг в данном месяце был получен вследствие очень малого количества отзывов.

Также можно заметить, что оба графика ведут себя достаточно схожим образом. То есть мы можем предположить, что есть некая корреляция между количеством отзывов в месяце и средним значением рейтинга.

```python
plt.title('scatter plot of average_rating and reviews_count')
sns.regplot(x='average_rating', y='reviews_count', data=monthly_data)
None
```

Выдвинем гипотезы $$H_0 = \{\text{истинная корреляция между этими величинами равна нулю}\},$$ $$H_1 = \overline{H_0}.$$
Для ее проверки воспользуемся статистическим критерием Спирмена.

```python
rho, p = spearmanr(monthly_data.average_rating, monthly_data.reviews_count)
print('spearman P-value: ' + str(p))
print('spearman correlation coefficient: ' + str(rho))
```

Взяв в качестве порогового значения $\epsilon = 0.05$, мы получим, что гипотеза $H_0$ отклоняется в силу альтернативной гипотезы. Таким образом, мы имеем право считать, что между этими величинами действительно присутствует корреляция, значение которой мы считаем по Спирмену равной 0.42.

В итоге можно утверждать, что количество отзывов действительно имеет линейную взаимосвязь со средним рейтингом в месяце.

# Natural Language Processing

Применим методы обработки естественного языка для дополнительного текстового анализа отзывов. Разделим отзывы на 2 группы: "положительные отзывы" с рейтингом 4-5 и "отрицательные отзывы" с рейтингом 1-3. Вычислим частоту встречающихся слов в обеих группах и выведем самые частые 20 слов для положительных отзывов и для отрицательных отзывов.

```python
from nltk.tokenize import RegexpTokenizer
from nltk.probability import FreqDist
import nltk
nltk.download('punkt')
nltk.download('wordnet')

tokenizer = RegexpTokenizer(r'\w+')

fdist_pos = FreqDist()
fdist_neg = FreqDist()

for review, rating in zip(df.review, df.rating):
    words = tokenizer.tokenize(review.lower())
    if rating > 3:
        for word in words:
            fdist_pos[word] += 1
    else:
        for word in words:
            fdist_neg[word] += 1
```

```python
fdist_pos.plot(20, title = 'frequency of top 20 words in positive reviews')
plt.show()
```

```python
fdist_neg.plot(20, title = 'frequency of top 20 words in negative reviews')
plt.show()
```

Такой анализ дает мало результатов. Можно заметить, что в обеих группах слова схожи. Однако в положительных отзывах также часто встречаются слова "kids", "great", "love" в отличе от отрицательных отзывов. В отрицательных отзывах, вероятно, чаще всего встречаются проблемы с картой и аккаунтом.

Попробуем составить биграммы слов и аналогично вычислить их частоту.

```python
fdist_bigram_pos = FreqDist()
fdist_bigram_neg = FreqDist()

for review, rating in zip(df.review, df.rating):
    tokens = tokenizer.tokenize(review.lower())
    bgs = nltk.bigrams(tokens)
    if rating > 2:
        for k, v in nltk.FreqDist(bgs).items():
            fdist_bigram_pos[k] += v
    else:
        for k, v in nltk.FreqDist(bgs).items():
            fdist_bigram_neg[k] += v
```

```python
fdist_bigram_pos.plot(20, title = 'frequency of top 20 bigrams in positive reviews')
plt.show()
```

Как можно видеть, среди положительных отзывов часто встречаются биграммы связанные с банковским приложением и детьми. Таким образом, наиболее частыми преимуществами являются удобства приложения и простота пользования для детей.

```python
fdist_bigram_neg.plot(20, title = 'frequency of top 20 bigrams in negative reviews')
plt.show()
```

Итак, среди негативных отзывов самая частая биграмма -- "the card", что свидетельствует о том, что пользователи, ставящие низкий рейтинг, испытывают некоторые проблемы с картами этого банка. Также одна из частых биграмм -- "customer service", что свидетельствует о проблемах с обслуживанием пользователей. Присутствуют биграммы связанные с детьми (как и в положительных отзывах), то есть присутствуют некоторые проблемы, связанные с использованием карт детьми. Как и в униграммах, в биграммах заметно появление слова "account", что свидетельствует о столкновении проблем с аккаунтом у пользователей.
